```python
%pylab inline
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Problema 1 Deformation 

# From class we know that the relationship between surface displacements and pressure change for a spherical source
#in an elastic half-space from Mogi(1958) is defined by the next parameters:

# a is the magma chamber radius (500-1000 m) but we are going to used in Km
a=linspace(0.5,1,num=50) 
# P is going to be the delta of pressure in the chamber (10-40 MPa)
P=linspace(10,40,num=50)
#G is the elastic shear modulus or rigidity (30 GPa)
G=linspace(3*10**4,1*10**5,num=50)
# d is the depth to the center of the magma chamber (3-10 km), but from the homework we used 2 - 40 km
d=linspace(2,40,num=50)
# r is the radial distance from the point of source (0-50 km)
r=linspace(0,50,num=50)

```

```python
# We are going to create the Moggi function for convenience 
def Moggi_function(a,P,G,d,r):
    Uz=3*(a**3)*P*d/(4*G*(((d**2)+(r**2))**(1.5)))
    Ur=3*(a**3)*P*r/(4*G*(((d**2)+(r**2))**(1.5)))
    return Uz,Ur
```

```python
# Define the maximun of each one
U1=Moggi_function(a[25],P,G[25],d[25],0)
Uz1= U1[0]
Ur1= U1[1]
maxUz= max(Uz1)
maxUr= max(Ur1)
```

```python
#Tested that the function max is ok and must print that for Ur the max is equal to 0. because the derivate of the function is 
# 
print (maxUz,maxUr)
```

```python
# define function for the graphs
def UZ (a,P,G,d,r):
    UZ=3*(a**3)*P*d/(4*G*(((d**2)+(r**2))**(1.5)))
    return UZ
def UR (a,P,G,d,r):
    UR=3*(a**3)*P*r/(4*G*(((d**2)+(r**2))**(1.5)))
    return UR
```

```python
# We plot the Max value for Uz for each parameter knowing that the maximun is when r=0
plt.figure(figsize=(20,20))
plt.plot(a,UZ(a,P[25],G[25],d[25],0))
plt.subplot(2,2,1)
xlabel('radius of the source a (km)',fontsize=15)
ylabel('MaxUz (cm)',fontsize=15)

plt.plot(P,UZ(a[25],P,G[25],d[25],0))
plt.subplot(2,2,2)
xlabel('delta of pressure P (MPa)',fontsize=15)
ylabel('MaxUz (cm)',fontsize=15)


plt.plot(G,UZ(a[25],P[25],G,d[25],0))
plt.subplot(2,2,3)
xlabel(' the elastic shear modulus G (MPa)',fontsize=15)
ylabel('MaxUz (cm)',fontsize=15)


plt.plot(d,UZ(a[25],P[25],G[25],d,0))
plt.subplot(2,2,4)
xlabel('d (Km)',fontsize=15)
ylabel('MaxUz (cm)',fontsize=15)
```

```python
#Function for UZ if r=0
def UZd(a,P,G,d):
    D= 3*(a**3)*P/(4*G*(d**2))
    return D
def URd (a,P,G,d):
    Dr= (3**0.5)*(a**3)*P/(6*G*(d**2))
    return Dr
plt.plot(d,UZd(a[25],P[25],G[25],d))
plt.subplot(2,1,1)
xlabel('depth d (Km)',fontsize=15)
ylabel('MaxUz (cm)',fontsize=15)
plt.plot(d,URd(a[25],P[25],G[25],d))
plt.subplot(2,1,2)
```

```python
 #Problema 3  3D surface deformation
# theta= 23
def R(x,y,f):
    R1=((x**2)+(y**2)+(f**2))**0.5
    return R1
def Ux (a,P,G,x,R):
    Ux1=3*(a**3)*P*x/(4*G*(R**3))
    return Ux1
def Uy (a,P,G,y,R):
    Uy1=3*(a**3)*P*y/(4*G*(R**3))
    return Uy1
def Uz (a,P,G,f,R):
    Uz1=3*(a**3)*P*f/(4*G*(R**3))
    return Uz1
#Define the function for Uz including the angle theta for the satellite

def Uzt (Uz,theta):
    Uzt1=Uz* cos(radians(theta))
    return Uzt1
```
